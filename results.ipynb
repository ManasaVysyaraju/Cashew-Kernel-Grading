```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import confusion_matrix
# Load the trained model
model = load_model('my_model.h5')

# Directory of test images
test_dir = 'test1'
def get_true_class(filename):
    if(filename[:4]=='W180'):
        return 0
    elif(filename[:4]=='W320'):
        return 1
    else:
        return 2
    
# Function to load and preprocess image for the model
def load_and_preprocess_image(img_path, target_size=(224, 224)):
    img = image.load_img(img_path, target_size=target_size)
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize to [0, 1]
    return img

# Lists to store true and predicted classes
true_classes = []
predicted_classes = []

# Iterate over all test images
for img_filename in os.listdir(test_dir):
    img_path = os.path.join(test_dir, img_filename)
    true_class = get_true_class(img_filename)
    true_classes.append(true_class)
    
    # Load and preprocess the image
    img = load_and_preprocess_image(img_path)
    
    # Predict the class
    prediction = model.predict(img)
    predicted_class = np.argmax(prediction, axis=1)[0]  # Assuming the output layer uses softmax
    predicted_classes.append(predicted_class)

cm= confusion_matrix(true_classes, predicted_classes)
print(cm)
print("Accuracy:", accuracy_score(true_classes, predicted_classes))
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
from sklearn.metrics import accuracy_score

# Load the trained model
model = load_model('Downloads/my_model.h5')

# Directory of test images
test_dir = 'Desktop/Cashew Kernel Classification.v2i.multiclass/valid'

# Function to extract the true class from the image filename
def get_true_class(filename):
    if(filename[:4]=='W180'):
        return 0
    elif(filename[:4]=='W320'):
        return 1
    else:
        return 2
    
# Function to load and preprocess image for the model
def load_and_preprocess_image(img_path, target_size=(224, 224)):
    img = image.load_img(img_path, target_size=target_size)
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize to [0, 1]
    return img

# Lists to store true and predicted classes
true_classes = []
predicted_classes = []

# Iterate over all test images
for img_filename in os.listdir(test_dir):
    img_path = os.path.join(test_dir, img_filename)
    true_class = get_true_class(img_filename)
    true_classes.append(true_class)
    
    # Load and preprocess the image
    img = load_and_preprocess_image(img_path)
    
    # Predict the class
    prediction = model.predict(img)
    predicted_class = np.argmax(prediction, axis=1)[0]  # Assuming the output layer uses softmax
    predicted_classes.append(predicted_class)

# Calculate accuracy
accuracy = accuracy_score(true_classes, predicted_classes)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
